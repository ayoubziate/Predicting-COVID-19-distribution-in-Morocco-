# Importing basic libs


```python
! pip install Beautifulsoup4
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

# Testing URL

```python
url="https://fr.wikipedia.org/wiki/Pandémie_de_Covid-19_au_Maroc"
# Make a GET request to fetch the raw HTML content
html_content = requests.get(url).text
```

# BeautifulSoup for extracting table data

```python
soup = BeautifulSoup(html_content,  "html.parser")
gdp_table = soup.find("table", {"class": "wikitable"})
gdp_table_data = gdp_table.tbody.find_all("tr")
gdp_table_data[3].find_all("td")[0].time.attrs['datetime']
```

# Looping table for extracting numbers

```python
dicts = {}
k=1
for tr in gdp_table_data:
    td = tr.find_all('td')
    row = [i.text.rstrip("\n") for i in td]
    dicts[k] = row
    k+=1
dicts.pop(1)
live_data= pd.DataFrame(dicts).T
live_data.columns = ["Date", "Tests effectués", "Cas négatifs", "Cas positifs", "Cas décédés", "Cas rétablis", "Nouveaux cas"]
```

# Saving data

```python
live_data.to_csv (r'Data_coronavirus_Maroc.csv', index = False, header=True)
```
